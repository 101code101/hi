###Black jack game


```python
import gym
```

```python
env=gym.make('Blackjack-v1', natural=False, sab=False)
```

```python
print(env.action_space)
print(env.observation_space)
```

```python
#actual code to run blackjake game
for i in range(10):
    observation=env.reset()
    while True:
        print(observation)
        action=env.action_space.sample()
        observation,reward,done,info=env.step(action)
        print(observation,reward,done)
        if done:
            print("Game end Reward:",reward)
            print("You won!") if reward>0 else print("You lost!")
            break
        
```

###Cart pole game


```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
env_name = 'CartPole-v0'
env = gym.make(env_name)
```

```python
for episode in range(1, 11):
    score = 0
    state = env.reset()
    done = False

    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score += reward

    print('Episode:', episode, 'Score:', score,n_state)
env.close()
```

###Mountain car

```python
import numpy as np
import gym

# Define the environment name
ENV_NAME = 'MountainCar-v0'

# Create the environment
env = gym.make(ENV_NAME)

# Define the number of episodes and maximum timesteps per episode
num_episodes = 5
max_timesteps = 100

# Play with random actions (no training)
for episode in range(num_episodes):
    observation, info = env.reset()  # Reset environment and get initial state
    print(f"Starting Episode {episode + 1}")
    for t in range(max_timesteps):
        env.render()  # Render the environment

        # Display observation
        print(f"Step {t + 1}: Observation: {observation}")

        # Take a random action
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)

        # Check if the episode is finished
        if done:
            print(f"Episode {episode + 1} finished after {t + 1} timesteps\n")
            break
    else:
        print(f"Episode {episode + 1} reached the maximum timestep limit.\n")

# Close the environment after use
env.close()
```
